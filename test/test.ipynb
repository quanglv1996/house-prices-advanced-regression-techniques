```python
for i, col in enumerate(df_object):
    len_cate = len(df[col].unique())
    if len_cate>0:
        print('ID: {}| Type: {}| {}'.format(i, col, df[col].unique()))
```

```python
for i, col in enumerate(df_object):
    onehot = pd.get_dummies(data=df_object, columns=[col])
    print('Col: {}'.format(col))
    print(onehot.head(10))
```

```python
df_onhot = pd.DataFrame()
for i, col in enumerate(df_object):
    onehot = pd.get_dummies(df_object[col])
    if i == 0:
        df_onhot = onehot
    else:
    
        # print('Col: {}'.format(col))
        # print(onehot)
        df_onhot = df_onhot.join(onehot, rsuffix='_{}'.format(i))

print(df_onhot)
```

```python
for i, col in enumerate(df_onhot):
    # print(col)
    count = df_onhot[col].isna().sum()
    # print('xxxxxxxxxxxx',count)
    # print('yyyyyyyyyyyy',type(count))
    if count > 0:
        print('aaa')
        print('Col: {}| Count: {}'.format(col, count))
```

```python
for i, col in enumerate(df_object):
    count = df_object[col].isna().sum()
    if count > 0:
        print('Col: {}| Count: {}'.format(col, count))
```
