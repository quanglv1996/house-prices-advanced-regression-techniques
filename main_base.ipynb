```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
path_data_train = 'D:\house-prices-advanced-regression-techniques\data/train.csv'
data_train = pd.read_csv(path_data_train)
data_train = data_train.set_index('Id')
```

```python
path_data_test = 'D:\house-prices-advanced-regression-techniques\data/test.csv'
data_test = pd.read_csv(path_data_test)
data_test = data_test.set_index('Id')
```

```python
x_train = data_train.loc[:, data_train.columns != 'SalePrice']
y_train = data_train.loc[:, data_train.columns == 'SalePrice']
x_test = data_test
```

```python
data = pd.concat([x_train, x_test], axis=0)
data.head()
```

```python
data_num = data.select_dtypes(include='number')
data_cate = data.select_dtypes(exclude='number')
```

```python
def check_missing(df):
    for col in df:
        num_missing = df[col].isnull().sum()
        if num_missing > 0:
            print('Col: {}| Count: {}'.format(col, num_missing))
```

```python
def process_missing_number_data(data):
    for col in data:
        mean_col = data[col].mean()
        data[col] = data[col].replace(np.nan, mean_col)
    check_missing(data)
    return data
```

```python
data_num_missing = process_missing_number_data(data_num)
```

```python
def process_missing_cate_data(data):
    for col in data:
        values = data[col].mode().values
        data[col] = data[col].fillna(values[0])
    check_missing(data)
    return data
```

```python
data_cate_missing =  process_missing_cate_data(data_cate)
```

```python

def create_onehot_encoder(df_cate):
    onehot_encoder = OneHotEncoder()
    for col in data_cate:
        data_cate[col] = data_cate[col].astype('category')
    onehot_encoder.fit(df_cate)
    return onehot_encoder
```

```python
onehot_encoder = create_onehot_encoder(data_cate_missing)
```

```python
id_data_train = 1461
```

```python
x_train_num = data_num_missing.iloc[data_num_missing.index < id_data_train]
x_train_num
```

```python
x_test_num = data_num_missing.iloc[~(data_num_missing.index < id_data_train)]
x_test_num
```

```python
x_train_cate = data_cate_missing[data_cate_missing.index < id_data_train]
x_test_cate = data_cate_missing[data_cate_missing.index >= id_data_train]
```

```python
x_train_onehot = pd.DataFrame(onehot_encoder.transform(x_train_cate).toarray())
x_train_onehot
check_missing(x_train_onehot)
```

```python
x_test_onehot = pd.DataFrame(onehot_encoder.transform(x_test_cate).toarray())
x_test_onehot
check_missing(x_test_onehot)
```

```python
x_train_onehot.index.name = 'Id'
x_test_onehot.index.name = 'Id'
x_train_onehot
x_test_onehot
```

```python
x_train_onehot.index +=1
data_train_final = pd.concat([x_train_num, x_train_onehot], axis=1)
data_train_final
```

```python
x_test_onehot.index +=1461
```

```python
data_test_final = pd.concat([x_test_num, x_test_onehot], axis=1)
data_test_final
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, cross_val_predict


Input = [('scale', StandardScaler()), ('polinomial', PolynomialFeatures(degree=2)),('mode', LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(x_train_num, y_train)
```

```python

y_hat = pipe.predict(x_test_num)
```

```python
res = zip(list(np.arange(1461, 5000)), list([int(list(i)[0]) for i in y_hat]))
df = pd.DataFrame(res, columns=['Id', 'SalePrice'])
df.to_csv('res.csv', index=None)
```
